El modelo principal que elegimos para realizar la optimizacion es el Weibull, que se define mediante la siguiente funcion:

$u(t)=1-e^{-at}$

Posteriormente el resultado de nuestro modelo lo reemplazaremos en la funcion de costo que se define por:

$f(p)=\sum_{i=1}^{n}(y(p;x_i)-\hat{y}_i)^2$

Para graficar nuestro modelo en Python ocupamos los modulos de numpy y matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

Los modulos que utilizaremos para las representaciones graficas son las siguientes:

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.patches as mpatches
```

Es importante señalar que la optimizacion de la curva entregada por el modelo la realizaremos con el modulo **curve_fit**
de la libreria scipy


```python
from scipy.optimize import curve_fit
```

Nuestra variable t estara definida de manera secuencial en el tiempo, el cual partira con el valor  
0.1 hasta 1 y de este rango obtendremos una lista de 20 valores con la cual podremos graficar una funcion con una mejor curvatura.  
Luego cargaremos una serie de datos experimentales con los cuales mediremos la diferencia de lo pronosticado por el modelo principal a utilizar.

```python
t = np.linspace(0.1,1,20)
#datos experimentales
datos_exp = [0.2,0.3,0.45,0.55,0.6,0.7,0.75,0.8,0.8,0.8,0.8,0.8,0.8,0.8,0.8,0.8,0.8,0.8,0.8,0.8]
datos_y = np.array(datos_exp)
```

Creamos un arreglo **a = [ ]** el cual contendra los valores que utilizaremos para la medicion de la funcion costo.  
Para ello, mediante un ciclo, rellenaremos **a = [ ]** con parametros desde 1 hasta 5, con salto 0.01 para asi dar una mayor precision en la tarea de  
encontrar el parametro mas optimo entregado por la funcion costo.

```python
a = []
for i in np.arange(1, 5.01, 0.01):
    val = round(i,3)
    #print(a)
    a.append(val)
```

En la variable **resultados** de igual dimension que a = [ ], se guardaran todos los resultados de la sumatoria en la que se esta evaluando el modelo Weibull, en donde posteriormente aplicaremos la funcion **np.amin()** con el cual sabremos cual es el minimo valor entregado por la funcion costo, ademas del parametro con el cual se llega a ese resultado minimo.

```python
resultados = [0]*len(a)
for x in range(len(a)):
    sumatoria = 0
    for y in range(len(t)):
        #--------Sumatoria de la funcion costo con implementacion del modelo weibull-----
        sumatoria += (1-np.exp(-a[x]*t[y])-datos_exp[y])**2
        resultados[x] = sumatoria
    #print('la diferencia de datos para a = ',x+1,' es = ',np.amin(resultados[x]))

parametro_optimo = resultados.index(np.amin(resultados))
print('valor minimo de sum = ',np.amin(resultados)," con a = ",a[parametro_optimo])
```

El parametro obtenido anteriormente lo guardaremos en una variable, donde ademas asignamos otra variable b, las cuales seran necesarias para reemplazarlas en los modelos que utilizaremos a continuacion.

```python
a = a[parametro_optimo]
b = 0.5
print('a = ',a)
```

Tomamos en consideracion dos modelos(Weibull, Higuchi), en los cuales evaluaremos nuestros parametros y veremos cual es el que mas se asemeja a los datos experimentales que hemos cargado anteriormente.

```python
def Weibull(t,a):
    func = 1-np.exp(-(a*t))
    return func
def Higuchi(t,a,b):
    func2= a*t**(b)
    return func2
```

## Modelo Weibull.

En esta seccion de codigo definiremos en primer lugar las etiquetas de nuestra figura.  
Luego graficaremos las curvas de la funcion Weibull evaluadas en sus dos extremos y el punto mas optimo obtenido anteriormente.

```python
#definicion de la figura
green_patch = mpatches.Patch(color='green', label='a = 1')
orange_patch = mpatches.Patch(color='orange', label='a = 5')
red_patch = mpatches.Patch(color='red', label='Ajuste de optimización')
datos_weibull = np.array(Weibull(t, a))
plt.title("Modelo Weibull")

#Graficar las curvas
plt.scatter(t, datos_exp, label='Data')
plt.plot(t,Weibull(t,a),'-')
plt.plot(t,Weibull(t,1),'-')
plt.plot(t,Weibull(t,5),'-')
plt.legend(handles=[green_patch,orange_patch,red_patch])
#Seccion de Optimización de curva
popt, pcov = curve_fit(Weibull, t, datos_exp)
plt.plot(t, Weibull(t, *popt), 'r--')
plt.show()
print("Datos Weibull: \n ",datos_weibull)
```

En la **seccion de optimizacion de curva** entraremos a utilizar la funcion **curve_fit()** de la libreria Scipy, la cual se utiliza de la siguiente manera: curve_fit(Funcion, datos_x, datos_y). Como se puede apreciar, el ajuste de curva para el modelo Weibull con los datos experimentales(linea roja) es muy similar por no decir casi identico, a la curva generada por el parametro **a** mas optimo(linea azul). 

## Modelo Higuchi.

Relizamos los mismos pasos para graficar la funcion Higuchi

```python
#definicion de la figura
blue_patch = mpatches.Patch(color='dodgerblue', label='a = 2.77')
orange_patch = mpatches.Patch(color='orange', label='a = 5')
red_patch = mpatches.Patch(color='red', label='Ajuste de optimizacíon')
datos_higuchi = np.array(Higuchi(t, a, b))
plt.title("Modelo Higuchi")

#Graficar las curvas
plt.scatter(t, datos_exp, label='Data')
plt.plot(t,Higuchi(t,a,b),'-')
plt.plot(t,Higuchi(t,5,b),'-')
plt.legend(handles=[orange_patch,blue_patch,red_patch])

#Seccion de Optimización de curva
popt2, pcov = curve_fit(Higuchi, t, datos_exp)
plt.plot(t, Higuchi(t, *popt2), 'r--')
plt.show()
print("Datos Higuchi: \n ",datos_higuchi)
```

## Otro ejemplo de optimizacion: Rosenbrock

Se define la funcion Rosenbrock.

```python
B = 10
f = lambda x,y: (x-1)**2 + B*(y-x**2)**2;
```

```python
# Evaluar la funcion con los siguientes datos
X = np.arange(-2, 2, 0.15)
Y = np.arange(-1, 3, 0.15)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)
```

Las siguientes lineas de codigo es simplemente para para mostrar el plot de nuestra funcion y los datos,
ademas de ajustar los ejes de la imagen.

```python
figRos = plt.figure(figsize=(12, 7))
axRos = figRos.gca(projection='3d')
surf = axRos.plot_surface(X, Y, Z, cmap=cm.gist_heat_r,
                       linewidth=0, antialiased=False)
axRos.set_zlim(0, 200)
figRos.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
```

Implementacion de la gradiante de la funcion Rosenbrock, la cual se define de la siguiente forma: $\triangledown f=(\frac{2(x-1)-4b(y-x^{2})x}{2b(y-x^{2})})$

```python
df = lambda x,y: np.array([2*(x-1) - 4*b*(y - x**2)*x, \
                         2*b*(y-x**2)])
```

Optimizacion: 

```python
F = lambda X: f(X[0],X[1])
dF = lambda X: df(X[0],X[1])

x0 = np.array([-1.4,1.1])

fx = F(x0);
gx = dF(x0);
s = -gx;
```

Se definen los parametros de la figura y agregamos los atributos para contornear nuestra gradiante de Rosenbrock.

```python
plt.figure(figsize=(12, 7))
plt.contour(X,Y,Z,200)
ns = np.sqrt(s[0]**2+s[1]**2);
plt.plot([x0[0]],[x0[1]],marker='o',markersize=15, color ='r')
plt.arrow(x0[0],x0[1],s[0]/ns,s[1]/ns, head_width=0.2, head_length=0.1, fc='r', ec='r')
plt.show()
```
